# **3 - Explorando variáveis quantitativas com medidas de tendência central**

Com as medidas de tendência central podemos conhecer as variáveis que representam as notas e nos preparar para fazer uma análise mais aprofundada com os histogramas e boxplots.

Calcular medidas de tendência central das notas isoladamente e depois da média de notas de cada aluno.

```python
media_nota_matematica = dados.Nota_matematica.mean()
mediana_nota_matematica = dados.Nota_matematica.median()
moda_nota_matematica = dados.Nota_matematica.mode()
print("A média das notas de matemática dos estudantes é %s, enquanto a mediana é %s. Essa váriavel é bimodal, sendo as modas das notas %s." %(media_nota_matematica,mediana_nota_matematica,moda_nota_matematica))
```

```python
media_nota_leitura = dados.Nota_leitura.mean()
mediana_nota_leitura = dados.Nota_leitura.median()
moda_nota_leitura = dados.Nota_leitura.mode()
print("A média das notas de leitura dos estudantes é %s, enquanto a mediana é %s, sendo as modas das notas %s." %(media_nota_leitura,mediana_nota_leitura,moda_nota_leitura))
```

```python
media_nota_escrita = dados.Nota_escrita.mean()
mediana_nota_escrita = dados.Nota_escrita.median()
moda_nota_escrita = dados.Nota_escrita.mode()
print("A média das notas de leitura dos estudantes é %s, enquanto a mediana é %s, sendo as modas das notas %s." %(media_nota_escrita,mediana_nota_escrita,moda_nota_escrita))
```
